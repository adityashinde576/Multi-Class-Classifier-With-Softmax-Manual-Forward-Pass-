```python
import math
w1=[
    [0.2,-0.5],
    [0.1,0.3]
]
b1=[0.0,0.1]
w2=[
    [0.4,0.2],
    [-0.3,0.1],
    [0.1,-0.2]
]
b2=[0.05,-0.05,0.1]
def relu(z):
    return [max(0,v) for v in z]
def softmax(z):
    exps=[math.exp(v) for v in z]
    total=sum(exps)
    return [v/total for v in exps]
def cce_loss(y_true,y_pred):
    epsilon=1e-15
    return -sum(y_true[i]*math.log(y_pred[i]+epsilon) for i in range(len(y_true)))
def forward_pass(x):
    z1=[w1[0][0]*x[0]+w1[0][1]*x[1]+b1[0],w1[1][0]*x[0]+w1[1][1]*x[1]+b1[1]]
    a1=relu(z1)
    z2=[w2[0][0]*a1[0]+w2[0][1]*a1[1]+b2[0],w2[1][0]*a1[0]+w2[1][1]*a1[1]+b2[1],w2[2][0]*a1[0]+w2[2][1]*a1[1]+b2[2]]
    a2=softmax(z2)
    return z1,a1,z2,a2
X=[[1.0,2.0],[0.5,-1.0],[2.0,1.0]]
Y=[[1,0,0],[0,1,0],[0,0,1]]
total_loss=0
for i in range(len(X)):
    x=X[i]
    y=Y[i]
    z1,a1,z2,a2=forward_pass(x)
    predicted_class=a2.index(max(a2))
    actual_class=y.index(1)
    loss=cce_loss(y,a2)
    total_loss+=loss
    print("Input:",x)
    print("z1:",z1)
    print("a1:",a1)
    print("z2:",z2)
    print("a2:",a2)
    print("Pred:",predicted_class)
    print("Actual:",actual_class)
    print("Loss:",loss)
avg_loss=total_loss/len(X)
print("Average Loss:",avg_loss)
```
